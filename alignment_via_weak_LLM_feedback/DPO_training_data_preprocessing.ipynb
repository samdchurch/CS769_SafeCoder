```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split
```

```python
training_data_path = "data_train_val/train/"
validation_data_path = "data_train_val/val/"
```

```python
training_sec_desc_df = pd.read_json(training_data_path + "sec-desc.jsonl", lines=True)
training_sec_new_desc_df = pd.read_json(training_data_path + "sec-new-desc.jsonl", lines=True)
training_evol_df = pd.read_json(training_data_path + "evol.jsonl", lines=True)
training_imsys_df = pd.read_json(training_data_path + "lmsys.jsonl", lines=True)


```

```python
unlabeled_sec_desc, dpo_sec_desc = train_test_split(training_sec_desc_df, test_size=.25, random_state=42)
unlabeled_new_sec_desc, dpo_new_sec_desc = train_test_split(training_sec_new_desc_df, test_size=.25, random_state=42)
```

```python
unlabeled_sec_desc.to_json("unlabeled_sec_desc.jsonl", orient='records', lines=True)
unlabeled_new_sec_desc.to_json("unlabeled_new_sec_desc.jsonl", orient='records', lines=True)
```

```python
combined_data = pd.concat([dpo_sec_desc, dpo_new_sec_desc], ignore_index=True)
combined_data = combined_data.sample(frac=1).reset_index(drop=True)
```

```python
combined_data
```

```python
def format_row(row):
    return {
        "prompt": row["description"],
        "preferred_output": row["func_src_after"],
        "less_preferred_output": row["func_src_before"],
    }

formatted_data = combined_data.apply(format_row, axis=1)
```

```python
formatted_data
```

```python
# Save to JSONL
output_file = "dpo_training_data.jsonl"
with open(output_file, "w") as f:
    for entry in formatted_data:
        json.dump(entry, f)
        f.write("\n")

print(f"Data saved to {output_file}")
```
